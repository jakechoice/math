# 2. Eigenvalues & Eigenvectors
# 
## $ \blacksquare $  수학적 의미와 구하는 방법
## 선형변환(Linear Transformation) $ R=Ar $ 에서
## 변환행렬 A 가 n $\times$ n 정방행렬(square matrix)이고, $ r \neq 0 $ 일때 
## 선형변환은 $ R=Ar= \lambda r $ 되는 $ \lambda $ 와  $ r $ 을 가질 수 있다.
## 이 때 $ \lambda $ 를 고유값(eigenvalues)이라 하고, 
## $ \lambda $ 에 대응되는 $ r $ 을 고유벡터(eigenvectors)라 한다.
#

## $ \blacksquare $ How to find eigenvalues and eigenvectors 
## Step 1.    Linear Transformation( $ R \rightarrow r,\ r \neq 0 , A $ is square )
## $$  R = Ar $$
## Step 2.    Make a homogeneous form
## $$  R=Ar= \lambda r $$
## $$ (A- \lambda I)r =0 $$
## Step 3.    Find $\lambda $ to satisfy this.(Find eigenvalues)
## $$ det(A- \lambda I) = 0 $$
## Step 4.    Find $ r $ for each $ \lambda . $ (Find eigenvectors by substitution $ \lambda $ into Step 2)
## $$ \lambda \rightarrow Step 2 $$
## Step 5.    Find the physical meaning at its own problem with the eigenvalues and eigenvectors
#
     
## $ \blacksquare $ 물리적 의미(주응력, 고유진동수,column buckling, structural vibration)
## 기본적으로 선형변환 행렬은 배열된 수들의 뭉치 덩어리($ A $)이므로 하나의 단일 값으로 표현할 수 없다.
## 그러나 eigenvector 방향( $ r = eigenvector $) 에서는 단일 수 값(eigenvalue, $ \lambda $)으로 표현되고,
## $$  R=Ar= \lambda r \tag{1} $$
## $$ (A- \lambda I)r =0 \tag{2} $$
## 그 값들(eigenvalues, eigenvectors)에서는 (1), (2) 가 성립하고(당연히), 
## 물리적으로 그 문제 고유의 의미를 갖게 된다.
#
## Example  $ 2 \times 2 $ matrix to find Eigenvalues and Eigenvectors
# $$
\begin{pmatrix} 
X \\ 
Y  
\end{pmatrix}
=

\end{pmatrix}

$$

##  $$  R=Ar= \lambda r \ $$

# $$
\begin{pmatrix}

$$

## (3.2)를 풀어쓰면
# $$
\begin{align}
5x-2y & = \lambda x  \\
-2x + 2y & = \lambda y 
\end{align} \tag{3.3} 
$$

# $$
\begin{align}
 (5-\lambda)x - 2y  & = 0 \\
 -2x + (2-\lambda)y & = 0
\end{align} \tag{3.4} 
$$
#
## (3.3)은 homogeneous equations이라 한다.
## Homogeneous Linear equations이 해를 가지기 위한 조건 : 

## You have to kow the meaning of *'rank'* in a matrix

# $$

$$

##  $$  R=Ar= \lambda r \ \tag{3.6} $$
## $$ Ar - \lambda r = (A-\lambda I)r = 0 \tag{3.7} $$

# $$
(A-\lambda I)r =
\begin{pmatrix} 
(5-\lambda) & -2 \\ 
-2 &  (2 -\lambda )
\end{pmatrix}
\begin{pmatrix}
x \\
y
\end{pmatrix} = 0 \tag{3.8}
$$
# (3.8) 이 해를 갖는 경우(homogeneous equations have solutions in case of two)
# Rank = n, at $ (n \times n)$ matrix
# when $$ det(A - \lambda I) \neq 0  : (x=0, y=0) $$
# in general form $$ det(A - \lambda I) \neq 0 , r=(0,0,0,,,0) \tag{3.9.1} $$
# when $$ det(A - \lambda I)=0 , r= r_{\lambda_1} , r_{\lambda_2} , r_{\lambda_3}, ,, $$
# so, $$ 
det(A - \lambda I) = 
\begin{vmatrix}
 (5-\lambda) - 2     \\
 -2 + (2-\lambda) 
\end{vmatrix} = 0 \tag{3.9.2} $$
#
# (3.9.2) is called the characteristic equation of matrix A
# Matrix A 의 특성방정식 :  대각행렬에서 $ \lambda $ 를 빼고 determinant(행렬식)을 0 으로 놓는다
# $ n \times n $ 행렬의 특성방정식은 $\lambda $ 의 n차 방정식이 된다. 
#
# Eigenvalues(고유값) : $ \lambda $ 의 n차 방정식 해가 곧 eigenvalues(고유값)이 된다
# 실제로 3차 이상의 고차방정식을 푸는 것은 쉽지 않기 때문에 
## 컴퓨터에서 수치계산으로 근사해를 계산하여 얻는다
#
# Eigenvectors(고유벡터): 
# eigenvalues(고유값) 각각에 대응하는 $ R=Ar= \lambda r $ 을 만족하는 $ r $
# (3.9.2) 를 풀면
# $$ (5- \lambda)(2- \lambda) - 4 = \lambda^2 - 7 \lambda + 6 = (\lambda-6)(\lambda-1)=0 $$
# $$ eigenvalues \ \ \lambda =1 \ \ or \ \ \lambda=6 $$
#
# Eigenvectors : $ \lambda $ 를 (3.5) 또는 (3.8)에 대입하면
# $$ when \ \ \lambda = 1, \ \ 2x - y = 0 \ \ or \ \ x-0.5y = 0 $$

\end{pmatrix}

\end{pmatrix}

\end{pmatrix}

\end{pmatrix}
= 0 \tag{3.9.5}
$$

# n-r dimensional(2-1=1 dimensional = line) solution of vector space exists $ \rightarrow $  x-0.5y = 0

```python
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as sla
import numpy.linalg as nla

A1 = np.array([[5,-2],[-2,2]])
B1 = np.array([[-2,2,-3],[2,1,-6],[-1,-2,0]])
C1 = np.array([[7.3,0.2,-3.7],[-11.5,1.0,5.5],[17.7,1.8,-9.3]])
A = np.asmatrix(A1)
B = np.asmatrix(B1)
C = np.asmatrix(C1)
print(A)
print(B)
print(C)
```

```python
A_ei = sla.eig(A)
print(A_ei)
```

```python
A_eival, A_eivec = nla.eig(A)
print(f'eigenvalues : {A_eival}')
print(f'engenvectors: {A_eivec}')
```

```python
B_ei = numpy.linalg.eig(B)
print(B_ei)
```

```python
B_eival, B_eivec = sla.eig(B)
print(f'eigenvalues : {B_eival}')
print(f'engenvectors: {B_eivec}')
```

```python
x_set=numpy.arange(-20,20,0.1)
y1_set=[]
y2_set=[]

for x in x_set:
  y1=2*x
  y2=-x/2
  y1_set.append(y1)
  y2_set.append(y2)
    

pyplot.plot(x_set,y1_set,label='2x-y=0')
pyplot.plot(x_set,y2_set,label='x+2y=0')
pyplot.title('Eigenvectors at each eigenvalues')
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.grid(alpha=5,linestyle='--')
pyplot.legend()
pyplot.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
